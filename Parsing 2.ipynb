Задача - вынуть еще и возрастной рейтинг игры. Он хранится на страницах, сслылка на которые имеет следующий вид:

https://www.metacritic.com/game/playstation-4/persona-5-strikers

Приводим к нужному виду названия и платформы.

```python
meta['l_name'] = meta['name'].str.lower()
meta['l_name'] = meta['l_name'].str.replace(':', '')
meta['l_name'] = meta['l_name'].str.replace('.', '')
meta['l_name'] = meta['l_name'].str.replace("'", '')
meta['l_name'] = meta['l_name'].str.replace(" & ", '-')
meta['l_name'] = meta['l_name'].str.replace(' ', '-')
meta['l_platform'] = meta['platform'].str.lower()
meta['l_platform'] = meta['l_platform'].str.replace(' ', '-')
meta.head(10)
```

```python
meta['link'] = 'https://www.metacritic.com/game/' + meta['l_platform'] + '/' + meta['l_name']
link_list = meta['link'].unique().tolist()
link_list
```

Проверили несколько - все открывается. Готовим парсинг.

```python
URL = 'https://www.metacritic.com/game/playstation-3/mass-effect-3'
```

```python
HEADERS = {
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/111.0'
}
```

```python
def get_html(url, params=''):
    r = requests.get(url, headers=HEADERS, params=params)
    return r
```

```python
h = get_html(URL)
h.text
```

```python
soup = BS(h.text, 'html.parser')
soup.find('a', class_='hover_none').get_text(strip=True)
```

```python
soup.find('span', class_='platform').get_text(strip=True)
```

```python
soup.find('li', class_='summary_detail product_rating').find('span', class_='data').text
```

Как-то вот так... Можно запускать. Первые сто сперва.

```python
games = []

for url in link_list[0:100]:
    try:
        h = get_html(url)
    except:
        print('Error in link', url)
    soup = BS(h.text, 'html.parser')
    try:
        
        games.append(
            {
                'name': soup.find('a', class_='hover_none').get_text(strip=True),
                'platform': soup.find('span', class_='platform').get_text(strip=True),
                'rating': soup.find('li', class_='summary_detail product_rating').find('span', class_='data').text
            })
    except:
        print('No rating in', url)
```

```python
def save_doc(items, path):
    with open(path, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['name', 'platform', 'rating'])
        for item in items:
            writer.writerow([item['name'], item['platform'], item['rating']])
```

```python
save_doc(games, r'D:\games\rat.csv')
```
