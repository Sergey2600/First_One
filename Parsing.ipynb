```python
URL = 'https://www.metacritic.com/browse/games/score/metascore/year/all/filtered?year_selected=2012&distribution=&sort=desc&view=detailed'
```

```python
HEADERS = {
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/111.0'
}
```

```python
def get_html(url, params=''):
    r = requests.get(url, headers=HEADERS, params=params)
    return r
```

```python
h = get_html(URL)
h.text
```

```python
games = []


def get_content(html):
    soup = BS(html, 'html.parser')
    items = soup.findAll('tr', class_='')
    
    for item in items:
        games.append(
        {
            'name': item.find('a', class_='title').find('h3').get_text(),
            'platform': item.find('span', class_='data').get_text(strip=True),
            'c_score': item.find('div', class_='clamp-metascore').find('div', class_='metascore_w').get_text(),
            'u_score': item.find('div', class_='clamp-userscore').find('div', class_='metascore_w').get_text()
        })
        
    return games
```

```python
C20120 = get_content(h.text)
C20120
```

Функция записи данных:

```python
def save_doc(items, path):
    with open(path, 'w', newline='') as file:
        writer = csv.writer(file, delimiter=':')
        writer.writerow(['name', 'platform', 'c_score', 'u_score'])
        for item in items:
            writer.writerow([item['name'], item['platform'], item['c_score'], item['u_score']])
```
